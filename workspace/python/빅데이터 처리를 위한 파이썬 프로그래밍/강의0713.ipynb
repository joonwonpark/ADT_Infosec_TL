```python
from bs4 import BeautifulSoup
import urllib.request as req

url = "https://movie.naver.com/movie/sdb/rank/rmovie.nhn?sel=cur&date=20210712"

response = req.urlopen(url)
soup = BeautifulSoup(response, "html.parser")
```

```python
soup.find_all('td', {'class' : 'title'})
```

```python
titleOfMovie = soup.find_all('div', {'class' : 'tit5'})
```

```python
pointOfMoive = soup.find_all('td', {'class' : 'point'})
```

```python
rank = 0
for title, point in zip(titleOfMovie, pointOfMoive):
    rank += 1
    print(rank, title.text.strip(), point.text)
```

```python
import requests

url = "https://www.weather.go.kr/plus/land/current/city.jsp"
response = requests.get(url) # <Response [200]> : 사이트 가져오기 성공
                              # 404 : url에 해당되는 사이트가 없다 ex) 오타, 없는 사이트
                              # 500, 505 : 서버 다운
```

```python
response.content
```

```python
response.text
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
```

```python
table = soup.find('table', {'class' : 'table_develop3'})
```

```python
for tr in table.find_all('tr'):
    try:
        print("=" * 100)
        print(tr.find('a').text) # 지점
        print(f'현재 온도는 {tr.find_all("td")[5].text}℃ 습도는{tr.find_all("td")[9].text}%')
    except:
        continue
    
```

```python
data = []
for tr in table.findAll("tr"):
  if tr.find('a'):
    data.append([tr.find("a").text,  tr.findAll("td")[5].text,  tr.findAll("td")[9].text ])
```

```python
with open("weather.csv", "w") as file:
    file.write("point,temperature,humidity\n")
```

```python
with open("weather.csv", "a") as file:
    for point, temp, humi in data:
        file.write(f"{point},{temp},{humi}\n")
```
