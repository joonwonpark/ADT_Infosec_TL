```python
X_train_onehot=[]
for seq in X_train:
    print("seq = ",  seq)
    input = [ num_dic[char] for char in seq ]
    print("input = ", input)
    onehot = np.eye(26)[input]
    print("*"*100)
    print("onehot")
    print("*"*100)
    print(onehot)
    X_train_onehot.append(onehot)
    print("="*100)
```

```python
y_train_tokens = [num_dic[char] for char in y_train]
```

```python
y_train_tokens
```

```python
np.eye(26)[y_train_tokens]
```

```python
y_train_onehot = np.eye(26)[y_train_tokens]
```

```python
from tensorflow.keras.models import Sequential
```

```python
model = Sequential()
```

image.png

```python
from tensorflow.keras.layers import SimpleRNN, Dense
```

```python
model.add(SimpleRNN( units=10, input_shape=(3, 26) ) )
```

image.png

```python
model.add(Dense(26, activation="softmax"))
```

```python
model.summary()
```

```python
from tensorflow.keras.optimizers import Adam
```

```python
model.compile(
                loss="categorical_crossentropy",
                optimizer=Adam(learning_rate=0.1),
                metrics=["acc"]
              )
```

```python
model.fit(X_train_onehot, y_train_onehot, epochs=30)
```
