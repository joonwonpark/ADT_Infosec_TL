```python
my_balance = 10000000000

def deposit(*messages, input_money = 200000000):
    global my_balance
    my_balance += input_money
    print("내 통장 잔액 = ", my_balance)
    print("messages = ", messages)  
    print("len(messages) = ", len(messages) )

    if len(messages) > 1:
        return messages[1]
    elif len(messages) > 0:
        return messages[0]
    else:
        return "김대리 임마 나도 칼퇴좀 해보자"
```

```python
my_balance = 10000000000

def withdraw(out_money):
    global my_balance
    
    if out_money > my_balance:
        return   # 오류나는 상황을 제외 시킬때
    
    my_balance -= out_money
    return out_money
```

```python
withdraw(1000000000000000)
```
