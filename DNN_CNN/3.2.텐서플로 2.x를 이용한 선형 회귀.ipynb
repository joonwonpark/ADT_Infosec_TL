```python
import numpy as np
```

```python
X = np.array(
      [
       1,
       2,
       3
       ], dtype="float32")
```

```python
X
```

```python
y = np.array([
              2, 
              2.5, 
              3.5
              ], dtype="float32")
```

```python
y
```

image.png

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Dense
```

```python
from tensorflow.keras.optimizers import Adam
```

```python
model = Sequential()
```

```python
model.add(Dense(1, input_dim=1))
```

image.png

```python
model.summary()
```

image.png

```python
model.compile(
    loss = "mean_squared_error",
    optimizer=Adam(lr=0.1)
    )
```

```python
model.fit(X, y , epochs=1000)
```

image.png
