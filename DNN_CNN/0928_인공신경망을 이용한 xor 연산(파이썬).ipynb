```python
import numpy as np
import tensorflow as tf
```

```python
X = np.array([
              [0,0],
              [0,1],
              [1,0],
              [1,1]
              ], dtype="float32")
```

```python
y = np.array([
              [0],
              [1],
              [1],
              [0]
            ], dtype="float32")
```

```python
w0 = tf.Variable([[1,2],[3,4]], dtype="float32")
```

```python
b0 = tf.Variable([[0,0]], dtype="float32")
```

```python
hypothesis0 = tf.sigmoid(tf.matmul(X, w0) + b0)
```

```python
hypothesis0
```

```python
w1 = tf.Variable([[5],[6]], dtype="float32")
```

```python
b1 = tf.Variable([[0]], dtype="float32")
```

```python
hypothesis1 = tf.sigmoid(tf.matmul(hypothesis0, w1) + b1)
```

```python
optimizer = tf.compat.v1.train.AdamOptimizer(learning_rate=0.1)
```

```python
for step in range(5000):
    with tf.GradientTape() as tape:
        hypothesis0 = tf.sigmoid(tf.matmul(X, w0) + b0)
        hypothesis1 = tf.sigmoid(tf.matmul(hypothesis0, w1) + b1)
        
        cost = -tf.reduce_mean(y * tf.math.log(hypothesis1) + (1 - y) * tf.math.log(1-hypothesis1))
        grads = tape.gradient(cost, [w0, w1, b0, b1])
        
        optimizer.apply_gradients(grads_and_vars = zip(grads, [w0, w1, b0, b1]))
        
        if step % 100 == 0:
            print(f"cost : {cost}, w0 : {w0}, w1 : {w1}")
        
```

# 심층 신경망을 이용한 XOR (tensorflow 2.x)

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import numpy as np
```

```python
X = np.array([
              [0,0],
              [0,1],
              [1,0],
              [1,1]
              ], dtype="float32")

y = np.array([0,1,1,0], dtype = 'float32')
```

```python
model = Sequential([
    Dense(2, input_dim = 2, activation = "sigmoid"),
    Dense(1, activation = "sigmoid")
])
```

```python
model.summary()
```

```python
model.compile(loss = 'binary_crossentropy', optimizer = Adam(learning_rate = 0.1), metrics=['acc'])
```

```python
model.fit(X, y, epochs = 500, verbose=0)
```

```python
pred = model.predict(X)
```

```python
pred
```

```python
predict01 = tf.cast(pred > 0.5, dtype=tf.float32)
```

```python
predict01 = tf.reshape(predict01, (1,-1))
```

```python
tf.reduce_mean(tf.cast(predict01 == y, dtype = tf.float32))
```
