```python
X = np.array([[0 , 0],
              [1 , 0],
              [0 , 0],
              [1 , 1]
              ], dtype = "float32")
```

```python
y = np.array([[0],
              [0],
              [0],
              [1]
             ], dtype = "float32")
```

image.png

```python
w = np.array([[1],
              [2]
              ], dtype="float32")
```

```python
b = np.array([3], dtype="float32")
```

```python
def sigmoid(hypothesis):
    return 1/(1 + np.exp(- hypothesis))
```

```python
learning_rate = 0.1
for i in range(10000):
    if i % 1000 == 0:
        print(f"{i}번째 학습")
        print(f"w = {w}")
        print(f"b = {b}")
        print(f"cost = {cost}")
    for j in range(2):
        hypothesis = sigmoid(np.dot(X, w) + b)
        cost = -1/4 * (y * np.log(hypothesis) + (1-y) * np.log(1 - hypothesis))
        x = X.T[j].reshape(1, -1)

        w_gred =  1/4 * np.sum(np.dot(x, (hypothesis - y)))

        b_gred = 1/4 * np.sum((hypothesis - y))

        w[j] = w[j] - learning_rate * w_gred
        b = b - learning_rate * b_gred
```

```python
predict = sigmoid(np.dot(X, w) + b)
```

```python
print(predict)
```

```python
predict_test = np.where(predict >= 0.5, 1, 0)
```

```python
acc = np.sum(predict_test == y) / 4
```

```python
print(acc)
```
