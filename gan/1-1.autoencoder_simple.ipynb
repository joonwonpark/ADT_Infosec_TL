```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Model
from tensorflow.keras.utils  import plot_model
```

```
# fashion_mnist.load_data() : fashion mnist 데이터를 읽어서 리턴
(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

```
#X_train.astype('float32')  :  X_train 이미지를 flat32 타입으로 변환
# 하고 255로 나눠줌
X_train = X_train.astype('float32') / 255.
```

```
#X_test.astype('float32')  :  X_test 이미지를 flat32 타입으로 변환
# 하고 255로 나눠줌
X_test = X_test.astype('float32') / 255.
```

```
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Flatten
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Reshape
from tensorflow.keras.losses import MeanSquaredError
from tensorflow.keras.optimizers import Adam
```

```
# 인코더 생성
# 28줄 28칸의 784개의 이미지를 입력 받아서 64개의 데이터로 축소 ( 이미지에서 중요한 특성만 남음)
encoder = Sequential([
              # 입력 데이터 28줄 28칸의 데이터를 1차원 배열 1줄 784칸으로 변환        
              Flatten(input_shape=(28, 28)),
              #입력 데이터를 64개로 축소
              Dense(64, activation="relu")        
])
```

```
encoder.summary()
```

```
# 인코더의 모양 조회
plot_model(encoder, show_dtype=True, show_shapes=True)
```

```
# 인코더가 축소한 64개의 데이터를 원래 모양인 28줄 28칸으로 복원
decoder = Sequential([
      # input_dim=64 : 64개의 데이터를 입력으로 (encoder가 축소한 데이터)
      # 784개의 데이터를 생성
      Dense(784, activation='sigmoid', input_dim=64),
      # 784개의 데이터를 28줄 28칸의 이미지로 복원
      Reshape((28, 28))
])
```

```
decoder.summary()
```

```
# 디코더의 모양 조회
plot_model(decoder, show_dtype=True, show_shapes=True)
```

```
class Autoencoder(Model):
    #객체 생성시 실행되는 함수
    #인코더와 디코더를 입력 받아서 속성에 저장  
    def __init__(self, encoder, decoder):
        super(Autoencoder, self).__init__()
        self.encoder = encoder
        self.decoder = decoder
    
    # 학습할때 실행되는 함수
    # 매개변수 x : 학습, 테스트 데이터
    def call(self, x):
        # 학습 또는 테스트 데이터를 encoder 연산 실행 (28줄 28칸의 784개 데이터를 64개로 줄임 => 이미지의 중요한 정보만 남음 )  
        encoded = self.encoder(x)
        # 인코딩한 이미지의 중요한 정보를 이용해서 원래 이미지로 복원 
        decoded = self.decoder(encoded)
        return decoded
```

```
# 오토 인코더 객체 생성
auto_encoder = Autoencoder(encoder, decoder) 
```

```
# 오토인코더 옵티마이저 loss 함수 설정
auto_encoder.compile(optimizer=Adam( learning_rate=0.001), loss=MeanSquaredError())
```

```
# 오토 인코더 학습
auto_encoder.fit(X_train, X_train,
                epochs=30,
                shuffle=True,
                validation_data=(X_test, X_test))
```

```
# X_test를 인코더를 이용해서 64개의 데이터로 압축
encoded_imgs = auto_encoder.encoder(X_test).numpy()
```

```
# 압축한 데이터를 원래 이미지로 복구
decoded_imgs = auto_encoder.decoder(encoded_imgs).numpy()
```
