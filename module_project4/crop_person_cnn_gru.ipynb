```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import sklearn
import matplotlib.pyplot as plt
import cv2
import cvlib as cv
from tqdm.notebook import tqdm
from cvlib.object_detection import draw_bbox
from google.colab.patches import cv2_imshow
import numpy as np
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split
```

```python
# 경로
csv_path = '/content/drive/MyDrive/모듈프로젝트4_team8/workspace/train.csv'

train_dir = '/content/drive/MyDrive/모듈프로젝트4_team8/workspace/train/'
```

```python
# 이미지 너비
IMG_WIDTH = 224

# 이미지 높이 
IMG_HEIGHT =224

# 비디오에서 학습할 프레임 개수
MAX_SEQ_LENGTH =20
```

```python
# bbox[1]_min : bbox[3]_max , bbox[0]_min : bbox[2]_max
def bbox_crop(frame,idx,bboxs):
    # idx : 몇번째 행 영상
    box = bboxs[idx]
    for i in range(2):
        if box[i] < 11:
            box[i] = 0
        else:
            box[i] -=10       
    return frame[box[1] : box[3]+10  , box[0] :box[2]+10 ]

# 리스트 슬라이싱
def list_slice(lst, n):
    return [lst[i:i+n] for i in range(0, len(lst), n)]

# 0뺴고 평균 구하기 - 소수점 날리기
def nan_mean(array_list):
    return round(np.mean([i for i in array_list if i>1]))
```

```python
def get_bbox(num, start):
    # 배팅볼 영상 118개 좌표
    num = num
    bboxs = [[[0] *4]] * num
    for i in tqdm(range(num)):
        plus_box = []
        cap =  cv2.VideoCapture(train_dir+train_df["video_name"].loc[i+start])
        ret, frame = cap.read()

        bbox, label, conf = cv.detect_common_objects(frame)

        for j in range(len(label)):
            if label[j] == 'person':
                plus_box.append(bbox[j])
                bboxs[i] = plus_box
    return bboxs

def concat_bbox(bboxs):
    # 좌표 하나로 합치기
    for i in range(len(bboxs)):
        bboxs_0 = []
        bboxs_1 = []
        bboxs_2 = []
        bboxs_3 = []
        #     # bbox[1]_min : bbox[3]_max , bbox[0]_min : bbox[2]_max

        for j in range(len(bboxs[i])):
            bboxs_0.append(bboxs[i][j][0])
            bboxs_1.append(bboxs[i][j][1])
            bboxs_2.append(bboxs[i][j][2])
            bboxs_3.append(bboxs[i][j][3])
        bboxs[i] = [min(bboxs_0), min(bboxs_1),  max(bboxs_2),max(bboxs_3) ]

    return bboxs

# 객체 검출 실패시 7개의 평균값으로 채움
def fill_mean(bboxs):
    box_t = np.array(bboxs).T
    for idx, arr in enumerate(box_t):

        # x1, y1, x2, y2로 나누기
        arr = box_t[idx].flatten()
        # print(arr)

        # 4개로 나눈걸  7개씩 짜르기
        arr_7_split = list_slice(arr,7)

        # 7개로 자른것중에 0번째
        zero_to_mean = arr_7_split[0]

        # 첫번째 널 값 채우기
        zero_to_mean = np.where(zero_to_mean > 0, zero_to_mean, nan_mean(zero_to_mean))

        for j,arr2 in enumerate(arr_7_split):

            # 0보다 작으면 7개의 평균값 반환
            arr2 = np.where(arr2 > 0, arr2, nan_mean(arr2))
            
            # 1번째부터 합치기
            if j > 0:
                zero_to_mean = np.concatenate((zero_to_mean, arr2), axis=None)

        # 값 수정
        box_t[idx] = zero_to_mean
    return box_t.T
```

```python
# 비디오 파일을 읽어서 각 프레임을 이미지로 변환해서 리턴
# path : 비디오 파일 경로
# max_frames : 이미지로 변환할 프레임수
# resize=(IMG_SIZE, IMG_SIZE) : 이미지의 가로 세로
def load_video(path, max_frames=20, resize=(IMG_HEIGHT, IMG_WIDTH)):

    # 비디오 파일을 읽어서 이미지로 변환 할 객페
    cap = cv2.VideoCapture(path)
    frames = []
    bbox_idx = 0
    try:
        while True:


            # cap.read() : 비디오를 읽어서 리턴
            # ret : 비디오 읽기가 성공 True, 더이상 읽을 비디오 프레임이 없으면 False
            # frame : 비디오 프레임 이미지를 리턴
            ret, frame = cap.read()

            # ret 가 False면 반복 종료
            if not ret:
                break

            # 비디오 이미지를 가운데 리턴
            frame = bbox_crop(frame,bbox_idx,bboxs)
            #비디오 이미지의 가로 세로를 resize=(224,224) 로 변환
            try:
                frame = cv2.resize(frame, resize)
            except Exception as e:
                print(str(e))
            # frame 은  [줄, 칸, B G R ] 로 구성되 있음 
            # frame [ : (모든줄), : (모든칸), R (인덱스2) G (인덱스1) B (인덱스0 )] 리턴
            frame = frame[:, :, [2, 1, 0]]
            # frame을 frames에 추가

            frames.append(frame)
            # frames에 저장된 데이터수가 max_frames와 같으면 종료
            if len(frames) == max_frames:
                bbox_idx +=1
                break
    finally:
        # 비디오 이미지 변환 종료
        cap.release()
    # frames를 numpy 배열로 변환 해서 리턴
    return np.array(frames)
```

```python
# 모든 비디오 파일의 이미지와 종류를 리턴
# video_name : 비디오 파일명
# label: 비디오 파일 종류
# root_dir : 비디오 파일 경로
def prepare_all_videos(video_name , label, root_dir):
    # 비디오 파일 이름의 개수
    num_samples = len(video_name)
    # video_name.values.tolist() : 비디오 파일 이름을 리스트로 변환 해서 리턴
    video_paths = video_name.values.tolist()
    # 비디오 종류를 리턴
    labels = label.values
    # 비디오 종류를 2차원 배열로 변환
    labels = labels.reshape(-1,1)
    # labels를 float32 로 변환
    labels = np.array(labels , dtype="float32")
    # 0으로 초기화된 [비디오 파일개수 , 20 , 224 , 224, 3] 배열 생성 전체 비디오 저장
    x = np.zeros(shape=(num_samples, MAX_SEQ_LENGTH, IMG_HEIGHT, IMG_WIDTH, 3 ), dtype="float32" )
    # 비디오 파일의 개수 만큼 반복
    for idx, path in enumerate(video_paths):
        # 0으로 초기화된 [ 20 , 224 , 224, 3] 배열 생성 비디오 1개 저장
        video = np.zeros(shape=(MAX_SEQ_LENGTH, IMG_HEIGHT, IMG_WIDTH, 3), dtype="float32")
        # 비디오 파일을 읽어서 각 프레임의 이미지를 리턴
        frames = load_video(root_dir+ path)
         # 프레임의 개수 만큼 반복
        for i in range(len(frames)):
             # frame의 i번째 이미지를 x에 추가
            video[i] = np.array(frames[i]/255, dtype="float32")
        # 비디오 1개의 정보를 x에 추가    
        x[idx] = video
    # 전체 비디오 프레임을 이미지로 변환한 x 
    # to_categorical(y) : 이미지의 종류 y 를 onehot 인코딩     
    return (x ,to_categorical(labels))
```

```python
# 모든 비디오 파일의 이미지와 종류를 리턴
# video_name : 비디오 파일명
# label: 비디오 파일 종류
# root_dir : 비디오 파일 경로
def prepare_all_videos(video_name , label, root_dir):
    # 비디오 파일 이름의 개수
    num_samples = len(video_name)
    # video_name.values.tolist() : 비디오 파일 이름을 리스트로 변환 해서 리턴
    video_paths = video_name.values.tolist()
    # 비디오 종류를 리턴
    labels = label.values
    # 비디오 종류를 2차원 배열로 변환
    labels = labels.reshape(-1,1)
    # labels를 float32 로 변환
    labels = np.array(labels , dtype="float32")
    # 0으로 초기화된 [비디오 파일개수 , 20 , 224 , 224, 3] 배열 생성 전체 비디오 저장
    x = np.zeros(shape=(num_samples, MAX_SEQ_LENGTH, IMG_HEIGHT, IMG_WIDTH, 3 ), dtype="float32" )
    # 비디오 파일의 개수 만큼 반복
    for idx, path in enumerate(video_paths):
        # 0으로 초기화된 [ 20 , 224 , 224, 3] 배열 생성 비디오 1개 저장
        video = np.zeros(shape=(MAX_SEQ_LENGTH, IMG_HEIGHT, IMG_WIDTH, 3), dtype="float32")
        # 비디오 파일을 읽어서 각 프레임의 이미지를 리턴
        frames = load_video(root_dir+ path)
         # 프레임의 개수 만큼 반복
        for i in range(len(frames)):
             # frame의 i번째 이미지를 x에 추가
            video[i] = np.array(frames[i]/255, dtype="float32")
        # 비디오 1개의 정보를 x에 추가    
        x[idx] = video
    # 전체 비디오 프레임을 이미지로 변환한 x 
    # to_categorical(y) : 이미지의 종류 y 를 onehot 인코딩     
    return (x ,to_categorical(labels))
```
