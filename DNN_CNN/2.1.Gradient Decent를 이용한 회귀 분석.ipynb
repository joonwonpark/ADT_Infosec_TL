```python
w = 2
b = 1

learning_rate = 0.1
```

```python
for i in range(10):
    print("w = ", w)
    print("b = ", b)
    hypothesis = X*w+b
    print("hypothesis = ", hypothesis)
    print("y = ", y)
    cost = np.sum((hypothesis -y)**2) /3
    print("cost = ", cost)

    w_gred = ( 1/3 * np.dot((hypothesis -y) , X) ) * learning_rate
    print("w_gred = ", w_gred)

    w = w - w_gred

    b_gred = (1/3 * np.sum(hypothesis - y) ) * learning_rate
    print("b_gred = ", b_gred)

    b = b - b_gred

    print("=" * 100)
```

```python
w = 2
b = 1

learning_rate = 0.1
```

```python
for i in range(100):
    print("w = ", w)
    print("b = ", b)
    hypothesis = X*w+b
    print("hypothesis = ", hypothesis)
    print("y = ", y)
    cost = np.sum((hypothesis -y)**2) /3
    print("cost = ", cost)

    w_gred = ( 1/3 * np.dot((hypothesis -y) , X) ) * learning_rate
    print("w_gred = ", w_gred)

    w = w - w_gred

    b_gred = (1/3 * np.sum(hypothesis - y) ) * learning_rate
    print("b_gred = ", b_gred)

    b = b - b_gred

    print("=" * 100)
```
