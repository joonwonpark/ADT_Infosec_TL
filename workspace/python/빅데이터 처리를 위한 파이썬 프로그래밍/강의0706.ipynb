```python
# 문자가 숫자로 자동으로 변신해서 저장됨 (Encoding)
# A -> 65
# B -> 66
str1 = "AB"
```

```python
# 문자가 숫자로 변신해서 저장 (Encoding)
# C -> 67
# D -> 68
str2 = "CB"
```

```python
str3 = "AD"
```

```python
# 맨 첫글자의 숫자를 비교해서 
# 맨첫글자의 숫자가 큰놈이 크다
# str1 => A (65) 시작
# str2 => C (68) 시작
# str2 가 크다
str1 < str2
```

```python
# str1 : "AB" => 65 66
# str3 : "AD" => 65  68
# 첫글자 숫자가 같으면 두번째 글자 숫자가 큰놈이 크다
str1 < str3
```

```python
last_character = number[-1]
```

```python
last_character
```

```python
int(last_character)
```

```python
last_number = int(last_character)
```

```python
last_number
```

```python
# 조건처럼 1줄로 써야되서 에러
if last_number == 0 
   or last_number == 2 
   or last_number == 4 :

    print("짝수 입니다")
```

```python
# 조건처럼 1줄로 써야는데 중간에 줄을 바꾸려면 
# \하고 줄 바꾸기
# \ 뒤에 공백 있으면 안됨
if last_number == 0 \
   or last_number == 2 \
   or last_number == 4 :

    print("짝수 입니다")
```
