```python
with tf.GradientTape() as tape:
    print("="*100)
    print("w0 = ", w0)
    print("b0 = ", b0)
    hypothesis0 = tf.sigmoid(tf.matmul(X, w0) + b0)
    print("hypothesis0 = ", hypothesis0)
    print("="*100)
    print("w1 =",w1)
    print("b1 = ", b1)
    hypothesis1 = tf.sigmoid(tf.matmul(hypothesis0, w1) + b1)
    print("hypothesis1 = ", hypothesis1 )
    cost = -tf.reduce_mean(y * tf.log(hypothesis1) + (1-y) * tf.log(1 - hypothesis1))
    print("cost = ", cost)
```

```python
with tf.GradientTape() as tape:
    print("="*100)
    print("w0 = ", w0)
    print("b0 = ", b0)
    hypothesis0 = tf.sigmoid(tf.matmul(X, w0) + b0)
    print("hypothesis0 = ", hypothesis0)
    print("="*100)
    print("w1 =",w1)
    print("b1 = ", b1)
    hypothesis1 = tf.sigmoid(tf.matmul(hypothesis0, w1) + b1)
    print("hypothesis1 = ", hypothesis1 )
    cost = -tf.reduce_mean(y * tf.log(hypothesis1) + (1-y) * tf.log(1 - hypothesis1))
    print("cost = ", cost)
    grads = tape.gradient(cost, [w0,w1,b0,b1])
    print("grads = ", grads)
```

```python
with tf.GradientTape() as tape:
    print("="*100)
    print("w0 = ", w0)
    print("b0 = ", b0)
    hypothesis0 = tf.sigmoid(tf.matmul(X, w0) + b0)
    print("hypothesis0 = ", hypothesis0)
    print("="*100)
    print("w1 =",w1)
    print("b1 = ", b1)
    hypothesis1 = tf.sigmoid(tf.matmul(hypothesis0, w1) + b1)
    print("hypothesis1 = ", hypothesis1 )
    print("="*100)
    cost = -tf.reduce_mean(y * tf.log(hypothesis1) + (1-y) * tf.log(1 - hypothesis1))
    print("cost = ", cost)
    grads = tape.gradient(cost, [w0,w1,b0,b1])
    print("grads = ", grads)
```

```python
optimizer = tf.train.AdamOptimizer(0.01)
```

```python
for step in range(5000):
    print("step = ", step)
    with tf.GradientTape() as tape:
        print("="*100)
        print("w0 = ", w0)
        print("b0 = ", b0)
        hypothesis0 = tf.sigmoid(tf.matmul(X, w0) + b0)
        print("hypothesis0 = ", hypothesis0)
        print("="*100)
        print("w1 =",w1)
        print("b1 = ", b1)
        hypothesis1 = tf.sigmoid(tf.matmul(hypothesis0, w1) + b1)
        print("hypothesis1 = ", hypothesis1 )
        print("="*100)
        cost = -tf.reduce_mean(y * tf.log(hypothesis1) + (1-y) * tf.log(1 - hypothesis1))
        print("cost = ", cost)
        grads = tape.gradient(cost, [w0,w1,b0,b1])
        print("grads = ", grads)
        optimizer.apply_gradients(grads_and_vars = zip(grads,[w0, w1, b0, b1]))
        print("*"*100)
        print("*"*100)
```
