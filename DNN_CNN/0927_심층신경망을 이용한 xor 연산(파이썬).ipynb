```python
import numpy as np
```

```python
X = np.array([
              [0,0],
              [0,1],
              [1,0],
              [1,1]
              ], dtype="float32")
```

```python
y = np.array([
              [0],
              [1],
              [1],
              [0]
            ], dtype="float32")
```

```python
w = np.array([
              [1],
              [2]
            ], dtype="float32")
```

```python
b = np.array([3], dtype="float32")
```

```python
def sigmoid(h):
    return 1 / (1 * np.exp(-h))
```

```python
hypothesis = sigmoid(np.dot(X, w) + b)
```

```python
learning_rate = 0.1
for i in range(1000):
    print("i = ", i)
    print("*" * 100)
    for j in range(2):
        print("j = ", j)
        hypothesis = sigmoid(np.dot(X,w)+b)
        print("hypothesis = ", hypothesis)

        cost = -1/4 * np.sum(y * np.log(hypothesis) + (1 - y)* np.log(1- hypothesis))
        print("cost = ", cost)
        X_col_2d = X.T[j].reshape(1, -1)
        print("x{} = {}".format(j+1, X_col_2d))
        w_gred = 1/4 * np.sum(np.dot ( X_col_2d , (hypothesis - y)))
        print("w{}_gred = {}".format(j+1, w_gred))
        b_gred = 1/4 * np.sum(hypothesis - y)
        print("b_gred = ", b_gred)

        w[j] = w[j] - learning_rate * w_gred
        print("w{} = {}".format(j+1, w[j]))
        b = b - learning_rate * b_gred
        print("b = ", b)
        print("="*100)
```
