```python
def print_numbers(*args):
    print(f"args = {args}")
    for arg in args:
        print("=" * 100)
        print(f"arg = {arg}")
        print("=" * 100)
        
    print("$" * 100)
```

```python
print_numbers(10, 20, 30, 40)
```

```python
y = [10, 20, 30, 40]
```

```python
print_numbers(*y) # 리스트 y의 데이터를 unpacking 해서 10, 20, 30, 40 => args에 대입
```

```python
print_numbers(y)
```

```python
def personal_info(name, age, address):
    print("name =",name)
    print("age = ", age)
    print("address = ", address)
```

```python
x = {"address":"부산", "name":"홍길순", "age":30}
```

```python
personal_info(**x)  # dict를 풀어해친다.(unpacking)
```

```python
personal_info(age=20, name="홍길동", address="서울")
```

```python
def personal_info(**kwargs):
    print(f"kwargs = {kwargs}")
    for item in kwargs:
        print("=" * 100)
        print(f"item = {item}")
        print("=" * 100)
        
    print("$" * 100)
```

```python
personal_info(hobby = "게임", first_name = "길동", last_name = "홍", national = "한국")
```

```python
def personal_info( **kwargs):
    print("kwargs = ", kwargs)
    for item in kwargs.items():
        print("=" * 100)
        print("item = ", item)
        print("=" * 100)

    print("$" * 100)    
```

```python
personal_info(hobby="게임", first_name="길동", last_name="홍", national="한국")
```

```python
y = {"job":"회사원", "company":"SK InfoSec", "salary":4000}
```

```python
personal_info(**y)
```
